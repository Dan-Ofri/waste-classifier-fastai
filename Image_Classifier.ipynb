```python
import shutil
from icrawler.builtin import BingImageCrawler
from pathlib import Path
from fastai.vision.all import *
```

```python


# 1. 
searches = ['crushed aluminum can', 'plastic bottle waste', 'cardboard box waste', 'glass bottle waste']
base_path = Path('recycling_dataset')

# 2.  
for search_term in searches:
    print(f"--- Starting collection for: {search_term} ---")
    
    #    (icrawler    ,  Path object)
    dest_dir = base_path / search_term
    
    #   (Crawler)
    # downloader_threads=4 ->  4   (!)
    bing_crawler = BingImageCrawler(
        downloader_threads=4,
        storage={'root_dir': str(dest_dir)},
        log_level='ERROR' #      
    )
    
    #    ( 100   )
    bing_crawler.crawl(keyword=search_term, max_num=100)
    
    print(f"Finished {search_term}. \n")

print(" Mission Accomplished! Check your folders.")
```

```python
path = Path('recycling_dataset')
for c in path.ls():
    print(f"{c.name}: {len(c.ls())} images")
```

```python
# 1.   ()
dblock = DataBlock(
    blocks=(ImageBlock, CategoryBlock),    # : , : 
    get_items=get_image_files,             #    ?   
    splitter=RandomSplitter(valid_pct=0.2, seed=42), #  20%   (Validation)
    get_y=parent_label,                    #     (parent folder)
    item_tfms=Resize(192)                  #       192x192
)

# 2.   (DataLoaders) 
path = Path('recycling_dataset')
dls = dblock.dataloaders(path)

# 3.   -    "" 
dls.show_batch(max_n=9)
```

```python
#   ()
learn = vision_learner(dls, resnet18, metrics=error_rate)

#   (Fine tuning)
learn.fine_tune(2)
```

```python
#    (Interpretation)
interp = ClassificationInterpretation.from_learner(learn)

#       (Loss  )
interp.plot_top_losses(9, figsize=(15,10))
```

```python
# 1.    (   )
recycling = DataBlock(
    blocks=(ImageBlock, CategoryBlock), 
    get_items=get_image_files, 
    splitter=RandomSplitter(valid_pct=0.2, seed=42),
    get_y=parent_label,
    item_tfms=Resize(192),
batch_tfms=aug_transforms(mult=2.0, do_flip=True, flip_vert=True)
)

# 2. 
dls = recycling.dataloaders(path)

# 3.    
dls.show_batch(max_n=6)
```

```python
# 1.    (  " ")
learn = vision_learner(dls, resnet34, metrics=error_rate)

# 2.   (6 )
learn.fine_tune(6)
```
